## 2. Prétraitement des Données

### Q6 : Sélection des features et séparation de la cible

```python
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']

df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())

X = df[features].values.astype(float) # shape: (891, 6)
y = df['Survived'].values.reshape(-1, 1)
```

```python
print(f"Dimensions de X : {X.shape}")
print(f"Dimensions de y : {y.shape}")
```

### Q7 : Imputation de l'Âge par la médiane

```python
# Calcul du nombre de lignes affectées avant imputation
lignes_manquantes = df['Age'].isnull().sum()
print(f"Nombre de lignes affectées par l'imputation : {lignes_manquantes}")

# Imputation par la médiane
valeur_mediane = df['Age'].median()
df['Age'] = df['Age'].fillna(valeur_mediane)

# Vérification
print(f"Valeurs manquantes après imputation : {df['Age'].isnull().sum()}")
```

### Q8 : Normalisation Min-Max

```python
# TODO Q8 : Normalisation Min-Max
X_min = X.min(axis=0)
X_max = X.max(axis=0)

# Application de la formule : (x - x_min) / (x_max - x_min)
denom = X_max - X_min
denom[denom == 0] = 1

X_norm = (X - X_min) / denom

# Vérification (optionnelle mais conseillée)
print(f"Min après normalisation : {X_norm.min()}") 
print(f"Max après normalisation : {X_norm.max()}") 
```

### Q9 : Train/Test Split manuel

```python
import numpy as np

np.random.seed(42) # Correspond au random_state=42 demandé
idx = np.random.permutation(len(X_norm))
split = int(0.8 * len(idx))

X_train, X_test = X_norm[idx[:split]], X_norm[idx[split:]]
y_train, y_test = y[idx[:split]], y[idx[split:]]

# Affichage des dimensions demandées
print(f"Dimensions X_train : {X_train.shape}, y_train : {y_train.shape}")
print(f"Dimensions X_test  : {X_test.shape}, y_test  : {y_test.shape}")
```

## 3. Architecture du Réseau et Propagation Avant

### Q10. Initialisation de Xavier/Glorot

```python
 class NeuralNetwork:
    def __init__(self, layer_dims): # ex: [6, 8, 8, 4, 1]
        self.L = len(layer_dims) - 1 # nombre de couches
        self.W, self.b = {}, {}

        print(f"{'Couche':<10} | {'Dimension W':<15} | {'Dimension b':<15}")
        print("-" * 45)
        for l in range(1, self.L + 1):
            n_in = layer_dims[l-1]
            n_out = layer_dims[l]
            # TODO Q10 : initialisation Xavier
            # Formule : W ~ randn * sqrt(2 / (n_in + n_out))
            self.W[l] = np.random.randn(n_out, n_in) * np.sqrt(2 / (n_in + n_out))
            self.b[l] = np.zeros((n_out, 1))
            
            print(f"Couche {l:<3} | {str(self.W[l].shape):<15} | {str(self.b[l].shape):<15}")

    def sigmoid(self, z): return 1 / (1 + np.exp(-z))
    def sigmoid_d(self, z): return self.sigmoid(z) * (1 - self.sigmoid(z))

    def forward(self, X): # X shape: (n_features, m_examples)
        self.Z, self.A = {}, {0: X}
        for l in range(1, self.L + 1):
        # TODO Q11 : Z[l] = W[l] @ A[l-1] + b[l]
            self.Z[l] = np.dot(self.W[l], self.A[l-1]) + self.b[l]
            self.A[l] = self.sigmoid(self.Z[l])
        return self.A[self.L]
```

#### Pour répondre à Q10 : Affichez les dimensions de chaque matrice de poids.

```python
nn = NeuralNetwork([6, 8, 8, 4, 1])
```

## 4. Fonction de Coût et Rétropropagation

### Q14. Fonction de Coût : Binary Cross-Entropy (BCE)

```python
def compute_cost(self, y_hat, y):
    m = y.shape[1]
    eps = 1e-8
    # Formule : J = -(1/m) * sum(y*log(y_hat) + (1-y)*log(1-y_hat))
    cost = -(1/m) * np.sum(y * np.log(y_hat + eps) + (1 - y) * np.log(1 - y_hat + eps))
    return float(cost)
```

### Q15. Rétropropagation (Backward Propagation)

```python
def backward(self, y):
    m = y.shape[1]
    self.dW, self.db = {}, {}
     # Couche de sortie
    delta = -(y - self.A[self.L]) # simplification BCE + sigmoid
    self.dW[self.L] = (1/m) * delta @ self.A[self.L-1].T
    self.db[self.L] = (1/m) * np.sum(delta, axis=1, keepdims=True)
    # Couches cachees (de L-1 jusqu'a 1)
    for l in range(self.L - 1, 0, -1):
    # TODO Q15 : retropropager delta
        delta = np.dot(self.W[l+1].T, delta) * self.sigmoid_d(self.Z[l])
        self.dW[l] = (1/m) * np.dot(delta, self.A[l-1].T)
        self.db[l] = (1/m) * np.sum(delta, axis=1, keepdims=True)
```
