## Cellule 0 — Setup, imports et génération des datasets

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.colors import ListedColormap, BoundaryNorm
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.svm import SVC
from sklearn.datasets import make_circles, make_moons, make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, classification_report,
    confusion_matrix, ConfusionMatrixDisplay
)
from sklearn.model_selection import (
    train_test_split, GridSearchCV, cross_val_score, learning_curve
)
from sklearn.pipeline import Pipeline

# ── Palette et style globaux ──────────────────────────────────────────────────
plt.rcParams.update({
    'figure.facecolor' : 'white',
    'axes.facecolor'   : '#F9FAFB',
    'axes.grid'        : True,
    'grid.alpha'       : 0.25,
    'grid.linestyle'   : '--',
    'font.family'      : 'DejaVu Sans',
    'font.size'        : 11,
    'axes.titlesize'   : 12,
    'axes.titleweight' : 'bold',
    'axes.spines.top'  : False,
    'axes.spines.right': False,
})

CMAP_BG  = ListedColormap(['#DBEAFE', '#FEE2E2'])  # fond bleu / rouge
COL_CLS  = ['#1D4ED8', '#B91C1C']                   # points bleu / rouge
SEED     = 42
np.random.seed(SEED)

# ── Grille de décision (réutilisée partout) ───────────────────────────────────
def make_grid(X, margin=0.6, steps=350):
    x_min, x_max = X[:,0].min()-margin, X[:,0].max()+margin
    y_min, y_max = X[:,1].min()-margin, X[:,1].max()+margin
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, steps),
                         np.linspace(y_min, y_max, steps))
    return xx, yy

def plot_boundary(ax, clf, X, y, xx, yy, title='', sv=True):
    """Trace frontière de décision + points + vecteurs support."""
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.28, cmap=CMAP_BG)
    ax.contour(xx, yy, Z, colors=['#374151'], linewidths=2.0)
    for cls, col in zip([0,1], COL_CLS):
        m = y == cls
        ax.scatter(X[m,0], X[m,1], c=col, s=22, alpha=0.80,
                   edgecolors='white', linewidths=0.4)
    if sv and hasattr(clf, 'support_vectors_'):
        sv_ = clf.support_vectors_
        ax.scatter(sv_[:,0], sv_[:,1], s=90, facecolors='none',
                   edgecolors='black', linewidths=1.6, zorder=5,
                   label=f'{len(sv_)} SV')
        ax.legend(fontsize=8, loc='upper right')
    ax.set_title(title, fontsize=10)
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())

# ── Génération des 3 datasets ─────────────────────────────────────────────────
datasets = {
    'Cercles': make_circles(n_samples=400, noise=0.09, factor=0.4, random_state=SEED),
    'Lunes'  : make_moons  (n_samples=400, noise=0.12,              random_state=SEED),
    'Blobs'  : make_blobs  (n_samples=400, centers=2, cluster_std=1.8, random_state=SEED),
}

# Normalisation + split pour chaque dataset
scaled = {}
for name, (X, y) in datasets.items():
    sc = StandardScaler()
    X_s = sc.fit_transform(X)
    Xtr, Xte, ytr, yte = train_test_split(X_s, y, test_size=0.25,
                                           stratify=y, random_state=SEED)
    scaled[name] = dict(X=X_s, y=y, Xtr=Xtr, Xte=Xte, ytr=ytr, yte=yte, scaler=sc)

# ── Aperçu des 3 datasets ─────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for ax, (name, d) in zip(axes, scaled.items()):
    for cls, col in zip([0,1], COL_CLS):
        m = d['y'] == cls
        ax.scatter(d['X'][m,0], d['X'][m,1], c=col, s=18, alpha=0.75,
                   edgecolors='white', lw=0.3, label=f'Classe {cls}')
    ax.set_title(f'Dataset : {name}  (n={len(d["y"])})')
    ax.legend(fontsize=9)
    ax.set_xlabel('x₁'); ax.set_ylabel('x₂')

plt.suptitle('Datasets utilisés dans ce TP (normalisés)', fontsize=13,
             fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('tp_fig0_datasets.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Setup terminé — 3 datasets prêts.')
```

## Partie 1 — Chaque type de noyau : comparaison + effets des hyperparamètres

### 1.1 — Les 4 noyaux fondamentaux côte à côte

On fixe tous les autres paramètres et on change uniquement le noyau.  
Objectif : voir comment la **forme de la frontière** change.

```python
# ── 4 noyaux × 3 datasets ─────────────────────────────────────────────────────
kernels_base = [
    ('linear',  {},                         'Linéaire'),
    ('poly',    {'degree':3,'coef0':1},      'Polynomial (d=3)'),
    ('rbf',     {'gamma':0.8},              'RBF (γ=0.8)'),
    ('sigmoid', {'gamma':0.3,'coef0':0.0},  'Sigmoïde'),
]

dataset_names = list(scaled.keys())
fig, axes = plt.subplots(4, 3, figsize=(15, 18))

scores_table = {}

for row, (ker, params, ker_name) in enumerate(kernels_base):
    for col, dname in enumerate(dataset_names):
        ax   = axes[row][col]
        d    = scaled[dname]
        clf  = SVC(kernel=ker, C=1.0, **params)
        clf.fit(d['Xtr'], d['ytr'])
        acc_tr = clf.score(d['Xtr'], d['ytr'])
        acc_te = clf.score(d['Xte'], d['yte'])
        scores_table[(ker_name, dname)] = (acc_tr, acc_te)

        xx, yy = make_grid(d['X'])
        plot_boundary(ax, clf, d['Xtr'], d['ytr'], xx, yy)

        emoji = '✅' if acc_te > 0.92 else '⚠️' if acc_te > 0.80 else '❌'
        ax.set_title(
            f'[{ker_name}] — {dname}\n'
            f'Train {acc_tr:.0%}  |  Test {acc_te:.0%}  {emoji}',
            fontsize=9)
        ax.set_xlabel('x₁', fontsize=8); ax.set_ylabel('x₂', fontsize=8)

plt.suptitle(
    '4 noyaux × 3 datasets — C=1.0 fixé, seul le noyau change\n'
    '(Cercles en haut = cas non linéaire difficile)',
    fontsize=13, fontweight='bold')
plt.tight_layout(rect=[0,0,1,0.97])
plt.savefig('tp_fig1_4kernels_3datasets.png', dpi=150, bbox_inches='tight')
plt.show()

# Tableau récapitulatif
print('\n' + '='*68)
print(f'{"Noyau":<22} {"Dataset":<12} {"Train":>8} {"Test":>8} {"Diagnostic":>15}')
print('='*68)
for (kn, dn), (tr, te) in scores_table.items():
    gap  = tr - te
    diag = '✅ Bon' if te>0.92 and gap<0.05 else '⚠️ Sur-apprent.' if gap>0.06 else '📉 Sous-apprent.' if te<0.80 else '~ Acceptable'
    print(f'{kn:<22} {dn:<12} {tr:>7.1%} {te:>7.1%}  {diag}')
print('='*68)
```

---
### 1.2 — Effet du paramètre C (tolérance aux erreurs)

**C contrôle le compromis biais / variance :**
- **C petit** → grande marge tolérante → frontière lisse → sous-apprentissage possible
- **C grand** → petite marge stricte → frontière complexe → surapprentissage possible

On travaille sur le dataset **Lunes** avec le noyau **RBF** (γ fixé = 1.0).

```python
# ── Effet de C — noyau RBF, γ=1.0, dataset Lunes ─────────────────────────────
C_values = [0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]
d = scaled['Lunes']
xx, yy = make_grid(d['X'])

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.ravel()

C_scores = []
for i, C in enumerate(C_values):
    clf = SVC(kernel='rbf', C=C, gamma=1.0)
    clf.fit(d['Xtr'], d['ytr'])
    acc_tr = clf.score(d['Xtr'], d['ytr'])
    acc_te = clf.score(d['Xte'], d['yte'])
    n_sv   = len(clf.support_vectors_)
    C_scores.append((C, acc_tr, acc_te, n_sv))

    # Couleur du titre selon état
    if acc_tr - acc_te > 0.07:
        t_col = '#DC2626'
        state = 'Surapprentissage ⚠️'
    elif acc_te < 0.82:
        t_col = '#D97706'
        state = 'Sous-apprentissage'
    else:
        t_col = '#059669'
        state = 'Bon équilibre ✅'

    plot_boundary(axes[i], clf, d['Xtr'], d['ytr'], xx, yy)
    axes[i].set_title(
        f'C = {C}\nTrain={acc_tr:.0%}  Test={acc_te:.0%}  SV={n_sv}\n{state}',
        color=t_col, fontsize=10)
    axes[i].set_xlabel('x₁'); axes[i].set_ylabel('x₂')

plt.suptitle('Effet de C sur le Kernel SVM — RBF (γ=1.0), Dataset Lunes\n'
             'C petit → marge large → frontière lisse | C grand → marge étroite → frontière complexe',
             fontsize=12, fontweight='bold')
plt.tight_layout(rect=[0,0,1,0.93])
plt.savefig('tp_fig2_effet_C.png', dpi=150, bbox_inches='tight')
plt.show()

# Courbe résumé
fig, axes2 = plt.subplots(1, 2, figsize=(13, 4))
Cs, trs, tes, svs = zip(*C_scores)

axes2[0].semilogx(Cs, [t*100 for t in trs], 'o-', color='#93C5FD', lw=2, ms=8, label='Train')
axes2[0].semilogx(Cs, [t*100 for t in tes], 's-', color='#1D4ED8', lw=2, ms=8, label='Test')
axes2[0].fill_between(Cs, [t*100 for t in trs], [t*100 for t in tes], alpha=0.12, color='#EF4444',
                       label='Gap (surapprentissage)')
axes2[0].set_xlabel('C (log)'); axes2[0].set_ylabel('Accuracy (%)')
axes2[0].set_title('Accuracy vs C'); axes2[0].legend(); axes2[0].set_ylim(50, 105)
axes2[0].axvline(1.0, color='gray', lw=1, linestyle=':', alpha=0.6)

axes2[1].semilogx(Cs, svs, '^-', color='#7C3AED', lw=2, ms=8)
axes2[1].set_xlabel('C (log)'); axes2[1].set_ylabel('Nombre de vecteurs support')
axes2[1].set_title('Vecteurs support vs C\n(+ de SV = frontière plus complexe)')
axes2[1].axvline(1.0, color='gray', lw=1, linestyle=':', alpha=0.6)

plt.suptitle('Résumé : impact de C', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('tp_fig2b_C_courbes.png', dpi=150, bbox_inches='tight')
plt.show()

print('\n📌 Lecture :  C optimal ≈ valeur où le gap Train/Test est minimal')
print('             Le nombre de SV diminue quand C augmente (marge plus petite)')
```

---
### 1.3 — Effet de γ sur le noyau RBF

$$K_{\text{RBF}}(x, x') = \exp(-\gamma \|x - x'\|^2)$$

**γ contrôle la portée du noyau :**
- **γ petit** → noyau large → chaque point influence une grande zone → frontière lisse
- **γ grand** → noyau étroit → chaque point n'influence que son voisinage direct → frontière très irrégulière

```python
# ── Effet de γ — noyau RBF, C=1.0, dataset Cercles ───────────────────────────
gamma_values = [0.01, 0.1, 0.5, 1.0, 5.0, 50.0]
d = scaled['Cercles']
xx, yy = make_grid(d['X'])

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.ravel()
g_scores = []

for i, gamma in enumerate(gamma_values):
    clf = SVC(kernel='rbf', C=1.0, gamma=gamma)
    clf.fit(d['Xtr'], d['ytr'])
    acc_tr = clf.score(d['Xtr'], d['ytr'])
    acc_te = clf.score(d['Xte'], d['yte'])
    n_sv   = len(clf.support_vectors_)
    g_scores.append((gamma, acc_tr, acc_te, n_sv))

    if gamma <= 0.1:
        state, t_col = 'Sous-apprentissage 📉', '#D97706'
    elif acc_tr - acc_te > 0.07:
        state, t_col = 'Surapprentissage ⚠️', '#DC2626'
    else:
        state, t_col = 'Équilibre ✅', '#059669'

    plot_boundary(axes[i], clf, d['Xtr'], d['ytr'], xx, yy)
    axes[i].set_title(
        f'γ = {gamma}\nTrain={acc_tr:.0%}  Test={acc_te:.0%}  SV={n_sv}\n{state}',
        color=t_col, fontsize=10)
    axes[i].set_xlabel('x₁'); axes[i].set_ylabel('x₂')

plt.suptitle('Effet de γ sur le Kernel SVM — RBF (C=1.0), Dataset Cercles\n'
             'γ petit → portée longue (global) | γ grand → portée courte (local)',
             fontsize=12, fontweight='bold')
plt.tight_layout(rect=[0,0,1,0.93])
plt.savefig('tp_fig3_effet_gamma.png', dpi=150, bbox_inches='tight')
plt.show()

# Courbe résumé
fig, ax = plt.subplots(figsize=(8, 4))
gs, trs, tes, svs = zip(*g_scores)
ax.semilogx(gs, [t*100 for t in trs], 'o-', color='#93C5FD', lw=2, ms=8, label='Train')
ax.semilogx(gs, [t*100 for t in tes], 's-', color='#1D4ED8', lw=2, ms=8, label='Test')
ax.fill_between(gs, [t*100 for t in trs], [t*100 for t in tes],
                alpha=0.12, color='#EF4444', label='Gap')
ax.set_xlabel('γ (log)'); ax.set_ylabel('Accuracy (%)')
ax.set_title('Accuracy vs γ — zone optimale entre 0.5 et 2'); ax.legend()
ax.set_ylim(40, 106)
ax.axvspan(0.5, 2.0, alpha=0.07, color='green', label='Zone optimale')
plt.tight_layout()
plt.savefig('tp_fig3b_gamma_courbe.png', dpi=150, bbox_inches='tight')
plt.show()
print('\n📌 La zone optimale de γ est celle où Train ≈ Test (gap minimal)')
```

---
### 1.4 — Effet du degré d sur le noyau Polynomial

$$K_{\text{poly}}(x, x') = (x^T x' + c)^d$$

Le degré **d** contrôle la complexité de la frontière polynomiale.

```python
# ── Effet du degré d — noyau Polynomial, C=1.0, coef0=1, dataset Cercles ──────
degrees = [1, 2, 3, 4, 6, 8]
d = scaled['Cercles']
xx, yy = make_grid(d['X'])

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.ravel()
d_scores = []

for i, deg in enumerate(degrees):
    clf = SVC(kernel='poly', degree=deg, coef0=1.0, C=1.0)
    clf.fit(d['Xtr'], d['ytr'])
    acc_tr = clf.score(d['Xtr'], d['ytr'])
    acc_te = clf.score(d['Xte'], d['yte'])
    n_sv   = len(clf.support_vectors_)
    d_scores.append((deg, acc_tr, acc_te, n_sv))

    if deg == 1:
        state, t_col = 'Linéaire (d=1) ≡ SVM Lin.', '#6B7280'
    elif acc_tr - acc_te > 0.07:
        state, t_col = 'Surapprentissage ⚠️', '#DC2626'
    elif acc_te < 0.82:
        state, t_col = 'Sous-apprentissage 📉', '#D97706'
    else:
        state, t_col = 'Bon ✅', '#059669'

    plot_boundary(axes[i], clf, d['Xtr'], d['ytr'], xx, yy)
    axes[i].set_title(
        f'Poly (d={deg})\nTrain={acc_tr:.0%}  Test={acc_te:.0%}  SV={n_sv}\n{state}',
        color=t_col, fontsize=10)
    axes[i].set_xlabel('x₁'); axes[i].set_ylabel('x₂')

plt.suptitle('Effet du degré d — Noyau Polynomial (C=1.0, coef₀=1), Dataset Cercles\n'
             'd=1 ≡ linéaire | d pair mieux adapté aux cercles',
             fontsize=12, fontweight='bold')
plt.tight_layout(rect=[0,0,1,0.93])
plt.savefig('tp_fig4_effet_degree.png', dpi=150, bbox_inches='tight')
plt.show()

# Tableau degrés
print('\n' + '='*55)
print(f'{"Degré":>8}  {"Train":>8}  {"Test":>8}  {"# SV":>8}  {"Gap":>8}')
print('='*55)
for deg, tr, te, sv in d_scores:
    print(f'{deg:>8}  {tr:>8.1%}  {te:>8.1%}  {sv:>8}  {tr-te:>8.1%}')
print('='*55)
print('\n📌 d impair → frontière asymétrique (mauvais pour les cercles)')
print('   d pair  → frontière symétrique (mieux adaptée)')
```

### 1.5 — Heatmap C × γ (GridSearch visuel)

On visualise l'interaction entre C et γ sur le dataset Cercles.

```python
# ── Heatmap C × γ ─────────────────────────────────────────────────────────────
C_grid     = [0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]
gamma_grid = [0.01, 0.1, 0.5, 1.0,  5.0,   20.0]
d = scaled['Cercles']

score_matrix_tr = np.zeros((len(C_grid), len(gamma_grid)))
score_matrix_te = np.zeros((len(C_grid), len(gamma_grid)))

for i, C in enumerate(C_grid):
    for j, g in enumerate(gamma_grid):
        clf = SVC(kernel='rbf', C=C, gamma=g)
        clf.fit(d['Xtr'], d['ytr'])
        score_matrix_tr[i,j] = clf.score(d['Xtr'], d['ytr'])
        score_matrix_te[i,j] = clf.score(d['Xte'], d['yte'])

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

for ax, mat, title in zip(
        axes,
        [score_matrix_tr, score_matrix_te],
        ['Accuracy TRAIN', 'Accuracy TEST ← à optimiser']):
    im = ax.imshow(mat, cmap='RdYlGn', vmin=0.5, vmax=1.0, aspect='auto')
    plt.colorbar(im, ax=ax, label='Accuracy')
    ax.set_xticks(range(len(gamma_grid))); ax.set_xticklabels(gamma_grid)
    ax.set_yticks(range(len(C_grid)));     ax.set_yticklabels(C_grid)
    ax.set_xlabel('γ (gamma)'); ax.set_ylabel('C')
    ax.set_title(title, fontsize=12)
    for i in range(len(C_grid)):
        for j in range(len(gamma_grid)):
            txt_col = 'white' if mat[i,j] > 0.90 or mat[i,j] < 0.65 else 'black'
            ax.text(j, i, f'{mat[i,j]:.2f}', ha='center', va='center',
                    fontsize=8.5, fontweight='bold', color=txt_col)

# Marquer le meilleur sur TEST
best_i, best_j = np.unravel_index(score_matrix_te.argmax(), score_matrix_te.shape)
axes[1].add_patch(plt.Rectangle((best_j-0.5, best_i-0.5), 1, 1,
                                  fill=False, edgecolor='#1D4ED8', lw=3.5))
axes[1].set_title(
    f'Accuracy TEST  ←  Meilleur : C={C_grid[best_i]}, γ={gamma_grid[best_j]} '
    f'({score_matrix_te[best_i,best_j]:.2%})', fontsize=11)

plt.suptitle('Heatmap Train / Test — Kernel RBF sur Cercles\n'
             'Comparer les deux : zone rouge sur TEST + verte sur TRAIN = surapprentissage',
             fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('tp_fig5_heatmap_CG.png', dpi=150, bbox_inches='tight')
plt.show()
print(f'\n📌 Meilleure combinaison sur TEST : C={C_grid[best_i]}, γ={gamma_grid[best_j]}')
print('   La zone "verte TRAIN + rouge TEST" révèle le surapprentissage.')
```

## Partie 2 — SVM avec et sans noyau : comparaison visuelle complète

On compare systématiquement le SVM **linéaire** (sans noyau) contre le SVM **RBF** (avec noyau)  
sur les 3 datasets, avec visualisation des frontières, des marges et des vecteurs support.

```python
# ── SVM linéaire vs SVM RBF — les 3 datasets ─────────────────────────────────
fig, axes = plt.subplots(3, 2, figsize=(13, 17))

configs = [
    ('SVM Linéaire\n(sans noyau)',   SVC(kernel='linear', C=1.0),          '#6366F1'),
    ('SVM RBF γ=0.8\n(avec noyau)',  SVC(kernel='rbf', C=1.0, gamma=0.8),  '#D97706'),
]

comparison_results = []

for row, dname in enumerate(dataset_names):
    d    = scaled[dname]
    xx, yy = make_grid(d['X'])
    for col, (label, clf, line_col) in enumerate(configs):
        ax = axes[row][col]
        clf.fit(d['Xtr'], d['ytr'])
        acc_tr = clf.score(d['Xtr'], d['ytr'])
        acc_te = clf.score(d['Xte'], d['yte'])
        n_sv   = len(clf.support_vectors_)
        comparison_results.append((dname, label.split('\n')[0], acc_tr, acc_te, n_sv))

        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
        ax.contourf(xx, yy, Z, alpha=0.25, cmap=CMAP_BG)
        ax.contour(xx, yy, Z, colors=[line_col], linewidths=2.5)

        # Tracer aussi la marge (decision_function)
        try:
            Zdf = clf.decision_function(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
            ax.contour(xx, yy, Zdf, levels=[-1, 1],
                       colors=[line_col], linewidths=1.0, linestyles='--', alpha=0.5)
        except Exception:
            pass

        for cls, col in zip([0,1], COL_CLS):
            m = d['Xtr'][:, 0] == d['Xtr'][d['ytr']==cls, 0].reshape(-1,1)
            m = d['ytr'] == cls
            ax.scatter(d['Xtr'][m,0], d['Xtr'][m,1], c=col,
                       s=22, alpha=0.80, edgecolors='white', lw=0.4)

        sv_ = clf.support_vectors_
        ax.scatter(sv_[:,0], sv_[:,1], s=110, facecolors='none',
                   edgecolors='black', linewidths=1.8, zorder=5, label=f'{n_sv} SV')

        emoji = '✅' if acc_te > 0.92 else '⚠️' if acc_te > 0.80 else '❌'
        ax.set_title(f'{label}  —  {dname}\n'
                     f'Train {acc_tr:.0%}  |  Test {acc_te:.0%}  {emoji}  |  {n_sv} SV',
                     fontsize=10, color=line_col)
        ax.legend(fontsize=8, loc='upper right')
        ax.set_xlabel('x₁'); ax.set_ylabel('x₂')
        ax.set_xlim(xx.min(), xx.max()); ax.set_ylim(yy.min(), yy.max())

plt.suptitle(
    'SVM Linéaire (gauche) vs SVM RBF (droite) sur 3 datasets\n'
    'Tirets = marges ±1 | Cercles noirs = vecteurs support',
    fontsize=13, fontweight='bold')
plt.tight_layout(rect=[0,0,1,0.96])
plt.savefig('tp_fig6_lin_vs_rbf.png', dpi=150, bbox_inches='tight')
plt.show()

# Tableau de synthèse
print('\n' + '='*72)
print(f'{"Dataset":<12} {"Modèle":<22} {"Train":>8} {"Test":>8} {"# SV":>8} {"Verdict":>12}')
print('='*72)
for dn, mn, tr, te, sv in comparison_results:
    v = '✅' if te>0.92 else '⚠️' if te>0.80 else '❌'
    print(f'{dn:<12} {mn:<22} {tr:>7.1%} {te:>7.1%} {sv:>8}  {v}')
print('='*72)
```

### 3.2 — Matrices de confusion : tous les modèles sur le dataset Cercles

```python
# ── Matrices de confusion ─────────────────────────────────────────────────────
models_cm = [
    ('SVM Linéaire',       SVC(kernel='linear', C=1.0)),
    ('SVM Poly (d=2)',     SVC(kernel='poly',   degree=2, coef0=1, C=1.0)),
    ('SVM Poly (d=4)',     SVC(kernel='poly',   degree=4, coef0=1, C=1.0)),
    ('SVM RBF (γ=0.1)',   SVC(kernel='rbf',    gamma=0.1,  C=1.0)),
    ('SVM RBF (γ=0.8)',   SVC(kernel='rbf',    gamma=0.8,  C=1.0)),
    ('SVM RBF (γ=20)',    SVC(kernel='rbf',    gamma=20.0, C=1.0)),
]

d = scaled['Cercles']
class_names = ['Classe 0 (intérieur)', 'Classe 1 (extérieur)']

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.ravel()

for i, (name, clf) in enumerate(models_cm):
    clf.fit(d['Xtr'], d['ytr'])
    y_pred = clf.predict(d['Xte'])
    acc    = accuracy_score(d['yte'], y_pred)
    cm     = confusion_matrix(d['yte'], y_pred)

    # Calcul manuel des métriques
    TP = cm[1,1]; TN = cm[0,0]; FP = cm[0,1]; FN = cm[1,0]
    prec    = TP/(TP+FP) if (TP+FP)>0 else 0
    recall  = TP/(TP+FN) if (TP+FN)>0 else 0
    f1      = 2*prec*recall/(prec+recall) if (prec+recall)>0 else 0

    disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                                   display_labels=['Cl.0', 'Cl.1'])
    disp.plot(ax=axes[i], cmap='Blues', colorbar=False)
    axes[i].set_title(
        f'{name}\nAcc={acc:.0%}  Prec={prec:.0%}  Rec={recall:.0%}  F1={f1:.0%}',
        fontsize=9)
    axes[i].set_xlabel('Prédit'); axes[i].set_ylabel('Réel')

plt.suptitle(
    'Matrices de confusion — Dataset Cercles (données test uniquement)\n'
    'Diagonale principale = bonnes prédictions | Hors-diagonale = erreurs',
    fontsize=12, fontweight='bold')
plt.tight_layout(rect=[0,0,1,0.93])
plt.savefig('tp_fig9_confusion_matrices.png', dpi=150, bbox_inches='tight')
plt.show()

# Rapport détaillé pour le meilleur modèle
print('\n=== Rapport détaillé — SVM RBF (γ=0.8) ===')
clf_best = SVC(kernel='rbf', gamma=0.8, C=1.0)
clf_best.fit(d['Xtr'], d['ytr'])
y_pred_best = clf_best.predict(d['Xte'])
print(classification_report(d['yte'], y_pred_best,
                             target_names=class_names))
```

### 3.3 — Courbes d'apprentissage

La **courbe d'apprentissage** montre comment le modèle se comporte en fonction de la taille des données.  
C'est l'outil diagnostic le plus fiable pour distinguer **sous-apprentissage**, **surapprentissage** et **bon modèle**.

```python
# ── Courbes d'apprentissage — 3 modèles à diagnostiquer ─────────────────────
d = scaled['Cercles']

models_lc = [
    ('SVM Linéaire\n(sous-apprentissage)',   SVC(kernel='linear', C=1.0),           '#6366F1'),
    ('SVM RBF γ=50, C=1000\n(surapprentissage)', SVC(kernel='rbf', gamma=50, C=1000),'#DC2626'),
    ('SVM RBF γ=0.8, C=1\n(bon modèle)',    SVC(kernel='rbf', gamma=0.8, C=1.0),   '#059669'),
]

train_sizes = np.linspace(0.10, 1.0, 12)
fig, axes = plt.subplots(1, 3, figsize=(17, 5))

for ax, (name, clf, color) in zip(axes, models_lc):
    ts, tr_scores, val_scores = learning_curve(
        clf, d['X'], d['y'],
        train_sizes=train_sizes, cv=5,
        scoring='accuracy', n_jobs=-1
    )
    tr_mean  = tr_scores.mean(axis=1)
    val_mean = val_scores.mean(axis=1)
    tr_std   = tr_scores.std(axis=1)
    val_std  = val_scores.std(axis=1)

    ax.plot(ts, tr_mean*100,  'o-', color=color,  lw=2, ms=6, label='Train')
    ax.plot(ts, val_mean*100, 's--', color='gray', lw=2, ms=6, label='Validation (CV5)')
    ax.fill_between(ts, (tr_mean-tr_std)*100,  (tr_mean+tr_std)*100,  alpha=0.12, color=color)
    ax.fill_between(ts, (val_mean-val_std)*100,(val_mean+val_std)*100, alpha=0.12, color='gray')

    gap_final = (tr_mean[-1] - val_mean[-1]) * 100
    ax.set_xlabel('Taille du training set'); ax.set_ylabel('Accuracy (%)')
    ax.set_title(f'{name}\nGap final = {gap_final:.1f}%', fontsize=10, color=color)
    ax.legend(fontsize=9); ax.set_ylim(40, 108)
    ax.axhline(95, color='black', lw=0.8, linestyle=':', alpha=0.4)

plt.suptitle(
    'Courbes d\'apprentissage — 3 diagnostics\n'
    'Sous-apprent. : les deux courbes basses | Sur-apprent. : grand gap | Bon : convergence proche',
    fontsize=12, fontweight='bold')
plt.tight_layout(rect=[0,0,1,0.90])
plt.savefig('tp_fig10_learning_curves.png', dpi=150, bbox_inches='tight')
plt.show()

print('\n📌 Lecture des courbes :')
print('  - Les deux courbes convergent BAS    → sous-apprentissage (biais élevé)')
print('  - Grand gap persistant               → surapprentissage (variance élevée)')
print('  - Les deux courbes convergent HAUT   → bon modèle')
```

### 3.4 — GridSearchCV avec validation croisée 5-fold

```python
# ── GridSearchCV — recherche exhaustive + validation croisée ──────────────────
d = scaled['Cercles']

param_grid = {
    'clf__kernel': ['linear', 'poly', 'rbf'],
    'clf__C'     : [0.1, 1.0, 10.0, 100.0],
    'clf__gamma' : [0.01, 0.1, 1.0, 5.0],   # ignoré pour linéaire
    'clf__degree': [2, 3, 4],                # ignoré sauf poly
}

pipe = Pipeline([('sc', StandardScaler()), ('clf', SVC())])
gs   = GridSearchCV(pipe, param_grid, cv=5, scoring='accuracy',
                    n_jobs=-1, refit=True, verbose=0)
gs.fit(d['Xtr'], d['ytr'])

print('=== RÉSULTATS GRIDSEARCHCV (CV=5) ===')
print(f'  Meilleurs paramètres   : {gs.best_params_}')
print(f'  Score CV (meilleur)    : {gs.best_score_:.4f}')
print(f'  Score TEST final       : {gs.score(d["Xte"], d["yte"]):.4f}')

# Top 10 combinaisons
results_cv = gs.cv_results_
top_idx = np.argsort(results_cv['mean_test_score'])[::-1][:10]
print('\n--- Top 10 combinaisons ---')
print(f'{"Rang":<6} {"Score CV":>10} {"Std":>8} {"kernel":<10} {"C":>8} {"gamma":>8} {"degree":>8}')
print('-'*65)
for rank, idx in enumerate(top_idx, 1):
    p = results_cv['params'][idx]
    print(f'{rank:<6} '
          f'{results_cv["mean_test_score"][idx]:>10.4f} '
          f'{results_cv["std_test_score"][idx]:>8.4f} '
          f'{p["clf__kernel"]:<10} '
          f'{p["clf__C"]:>8} '
          f'{p["clf__gamma"]:>8} '
          f'{p["clf__degree"]:>8}')

# Visualisation de la frontière optimale
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
xx, yy = make_grid(d['X'])

# Meilleur modèle
best_pipe = gs.best_estimator_
y_pred_best = best_pipe.predict(d['Xte'])

# Frontière dans l'espace normalisé
best_clf = best_pipe.named_steps['clf']
sc_ = best_pipe.named_steps['sc']

Z = best_clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
axes[0].contourf(xx, yy, Z, alpha=0.25, cmap=CMAP_BG)
axes[0].contour(xx, yy, Z, colors=['#059669'], linewidths=2.5)
for cls, col in zip([0,1], COL_CLS):
    m = d['ytr'] == cls
    axes[0].scatter(d['Xtr'][m,0], d['Xtr'][m,1], c=col, s=20, alpha=0.7)
    m = d['yte'] == cls
    axes[0].scatter(d['Xte'][m,0], d['Xte'][m,1], c=col, s=55, marker='*',
                    edgecolors='black', lw=0.5, alpha=0.95)
sv_ = best_clf.support_vectors_
axes[0].scatter(sv_[:,0], sv_[:,1], s=120, facecolors='none',
                edgecolors='black', lw=1.8, label=f'{len(sv_)} SV')
axes[0].set_title(
    f'Meilleur modèle (GridSearch)\n'
    f'{gs.best_params_["clf__kernel"]} | C={gs.best_params_["clf__C"]} | '
    f'Test={gs.score(d["Xte"], d["yte"]):.0%}',
    fontsize=10, color='#059669')
axes[0].legend(fontsize=9)

# Matrice de confusion du meilleur
cm_best = confusion_matrix(d['yte'], y_pred_best)
ConfusionMatrixDisplay(cm_best, display_labels=['Cl.0','Cl.1']).plot(
    ax=axes[1], cmap='Greens', colorbar=False)
acc_best = accuracy_score(d['yte'], y_pred_best)
axes[1].set_title(f'Confusion matrix — Meilleur modèle\nAccuracy test = {acc_best:.2%}', fontsize=11)

plt.suptitle('GridSearchCV — Résultat optimal', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('tp_fig11_gridsearch_best.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Partie 4 — Analyse critique

### 4.1 — Le piège du surapprentissage : Train parfait ≠ bon modèle

```python
# ── Illustration du surapprentissage progressif ────────────────────────────────
# On augmente volontairement γ et C jusqu'à mémorisation totale

d = scaled['Cercles']
params_overfit = [
    (0.5,   1.0,   'Équilibre'),
    (2.0,   10.0,  'Légèrement sur-ajusté'),
    (10.0,  100.0, 'Sur-ajustement modéré'),
    (50.0,  1000.0,'Sur-ajustement sévère ⚠️'),
    (200.0, 10000.0,'Mémorisation totale ❌'),
]

fig, axes = plt.subplots(1, 5, figsize=(22, 4.5))
xx, yy = make_grid(d['X'])

for ax, (gamma, C, label) in zip(axes, params_overfit):
    clf = SVC(kernel='rbf', gamma=gamma, C=C)
    clf.fit(d['Xtr'], d['ytr'])
    acc_tr = clf.score(d['Xtr'], d['ytr'])
    acc_te = clf.score(d['Xte'], d['yte'])
    n_sv   = len(clf.support_vectors_)

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.25, cmap=CMAP_BG)
    ax.contour(xx, yy, Z, colors=['#374151'], linewidths=1.5)
    for cls, col in zip([0,1], COL_CLS):
        m = d['ytr'] == cls
        ax.scatter(d['Xtr'][m,0], d['Xtr'][m,1], c=col, s=12, alpha=0.6)

    gap = acc_tr - acc_te
    t_col = '#059669' if gap < 0.03 else '#D97706' if gap < 0.10 else '#DC2626'
    ax.set_title(
        f'{label}\nγ={gamma}  C={C}\n'
        f'Train={acc_tr:.0%}  Test={acc_te:.0%}\n'
        f'Gap={gap:.0%}  SV={n_sv}',
        fontsize=8.5, color=t_col)
    ax.set_xlim(xx.min(), xx.max()); ax.set_ylim(yy.min(), yy.max())

plt.suptitle(
    'Progression du surapprentissage — Kernel RBF\n'
    'Train=100% avec mémorisation → Generalization collapse',
    fontsize=12, fontweight='bold')
plt.tight_layout(rect=[0,0,1,0.88])
plt.savefig('tp_fig12_overfit_progression.png', dpi=150, bbox_inches='tight')
plt.show()
print('\n📌 Paradoxe du surapprentissage :')
print('   Train=100% avec γ=200, C=10000 → le modèle mémorise les données d\'entraînement')
print('   mais ne généralise pas → Test dégradé')
```

---
### 4.2 — Comportement en haute dimension : quand le noyau RBF perd face au linéaire

```python
# ── RBF vs Linéaire en fonction du nombre de dimensions ───────────────────────
# Sur des données LINÉAIREMENT séparables : le linéaire doit gagner

dims       = [2, 5, 10, 20, 50, 100, 200, 500]
n_samples  = 300
n_reps     = 5  # répétitions pour réduire la variance

acc_rbf_mean, acc_lin_mean = [], []
acc_rbf_std,  acc_lin_std  = [], []

for d in dims:
    r_rbf, r_lin = [], []
    for seed in range(n_reps):
        X_hd, y_hd = make_blobs(n_samples=n_samples, centers=2,
                                 n_features=d, cluster_std=2.0, random_state=seed)
        Xtr, Xte, ytr, yte = train_test_split(X_hd, y_hd, test_size=0.3,
                                               stratify=y_hd, random_state=seed)
        sc = StandardScaler()
        Xtr_s = sc.fit_transform(Xtr)
        Xte_s = sc.transform(Xte)
        r_rbf.append(SVC(kernel='rbf',    gamma='scale', C=1.0).fit(Xtr_s,ytr).score(Xte_s,yte))
        r_lin.append(SVC(kernel='linear',               C=1.0).fit(Xtr_s,ytr).score(Xte_s,yte))
    acc_rbf_mean.append(np.mean(r_rbf)); acc_rbf_std.append(np.std(r_rbf))
    acc_lin_mean.append(np.mean(r_lin)); acc_lin_std.append(np.std(r_lin))

fig, ax = plt.subplots(figsize=(10, 5))
ax.errorbar(dims, [v*100 for v in acc_rbf_mean], yerr=[v*100 for v in acc_rbf_std],
            fmt='o-', color='#D97706', lw=2.5, ms=8, capsize=5, label='RBF (gamma=scale)')
ax.errorbar(dims, [v*100 for v in acc_lin_mean], yerr=[v*100 for v in acc_lin_std],
            fmt='s--', color='#1D4ED8', lw=2.5, ms=8, capsize=5, label='Linéaire')
ax.set_xscale('log')
ax.set_xlabel('Nombre de dimensions (log)', fontsize=12)
ax.set_ylabel('Accuracy test (%)', fontsize=12)
ax.set_title(
    'RBF vs Linéaire — Impact de la dimensionnalité\n'
    '(données linéairement séparables, n=300, moyenne sur 5 seeds)',
    fontsize=12, fontweight='bold')
ax.legend(fontsize=11); ax.set_ylim(50, 105)
ax.axhline(90, color='gray', lw=1.2, linestyle=':', alpha=0.6)
ax.fill_between(dims,
                [v*100-s*100 for v,s in zip(acc_rbf_mean, acc_rbf_std)],
                [v*100+s*100 for v,s in zip(acc_rbf_mean, acc_rbf_std)],
                alpha=0.08, color='#D97706')
ax.fill_between(dims,
                [v*100-s*100 for v,s in zip(acc_lin_mean, acc_lin_std)],
                [v*100+s*100 for v,s in zip(acc_lin_mean, acc_lin_std)],
                alpha=0.08, color='#1D4ED8')
plt.tight_layout()
plt.savefig('tp_fig13_haute_dimension.png', dpi=150, bbox_inches='tight')
plt.show()

print('\n📌 Résultat : en grande dimension sur données linéaires, le SVM linéaire')
print('   maintient sa performance tandis que RBF peut fluctuer davantage.')
print('   → Règle : si d >> n (plus de features que de samples), préférer linéaire.')
```

---
### 4.3 — Impact du bruit : robustesse des noyaux

```python
# ── Robustesse au bruit ────────────────────────────────────────────────────────
noise_levels = [0.0, 0.05, 0.10, 0.20, 0.35, 0.50]

results_noise = {k: [] for k in ['linear', 'poly_d3', 'rbf_tuned', 'rbf_overfit']}

for noise in noise_levels:
    X_n, y_n = make_circles(n_samples=400, noise=noise, factor=0.4, random_state=SEED)
    Xtr, Xte, ytr, yte = train_test_split(X_n, y_n, test_size=0.25,
                                           stratify=y_n, random_state=SEED)
    sc = StandardScaler()
    Xtr_s = sc.fit_transform(Xtr); Xte_s = sc.transform(Xte)

    for key, clf in [
        ('linear',     SVC(kernel='linear', C=1.0)),
        ('poly_d3',    SVC(kernel='poly', degree=3, coef0=1, C=1.0)),
        ('rbf_tuned',  SVC(kernel='rbf', gamma=0.8, C=1.0)),
        ('rbf_overfit',SVC(kernel='rbf', gamma=20, C=1000)),
    ]:
        clf.fit(Xtr_s, ytr)
        results_noise[key].append(clf.score(Xte_s, yte))

fig, ax = plt.subplots(figsize=(10, 5))
styles = {
    'linear'     : ('#6366F1', 's--', 'SVM Linéaire'),
    'poly_d3'    : ('#7C3AED', '^-',  'Poly (d=3)'),
    'rbf_tuned'  : ('#059669', 'o-',  'RBF (γ=0.8, C=1) — bien réglé'),
    'rbf_overfit': ('#DC2626', 'x-',  'RBF (γ=20, C=1000) — sur-ajusté'),
}
for key, (col, fmt, lbl) in styles.items():
    ax.plot(noise_levels, [v*100 for v in results_noise[key]],
            fmt, color=col, lw=2.5, ms=9, label=lbl)

ax.set_xlabel('Niveau de bruit (noise)', fontsize=12)
ax.set_ylabel('Accuracy test (%)', fontsize=12)
ax.set_title(
    'Robustesse au bruit des 4 modèles — Dataset Cercles\n'
    'Bien réglé = dégradation lente | Sur-ajusté = chute rapide',
    fontsize=12, fontweight='bold')
ax.legend(fontsize=10); ax.set_ylim(40, 105)
ax.axvspan(0.3, 0.5, alpha=0.06, color='red', label='Zone bruit élevé')
ax.axhline(80, color='gray', lw=1, linestyle=':', alpha=0.5)
plt.tight_layout()
plt.savefig('tp_fig14_bruit.png', dpi=150, bbox_inches='tight')
plt.show()

print('\n📌 Le modèle sur-ajusté (γ=20, C=1000) dégrade beaucoup plus vite')
print('   quand le bruit augmente → il apprend le bruit au lieu de la structure.')
print('   Un modèle bien réglé (γ=0.8, C=1) est plus robuste.')
```

---
### 4.4 — Scalabilité : temps de calcul en fonction de n

```python
# ── Scalabilité : temps d'entraînement vs n ───────────────────────────────────
import time

sample_sizes = [100, 200, 500, 1000, 2000, 4000]

times = {'linear': [], 'rbf': []}
for n in sample_sizes:
    X_sc, y_sc = make_circles(n_samples=n, noise=0.1, factor=0.4, random_state=SEED)
    sc = StandardScaler()
    X_sc_s = sc.fit_transform(X_sc)
    for ker in ['linear', 'rbf']:
        clf = SVC(kernel=ker, C=1.0, gamma=0.8 if ker=='rbf' else 'auto')
        t0 = time.perf_counter()
        clf.fit(X_sc_s, y_sc)
        times[ker].append(time.perf_counter() - t0)

fig, ax = plt.subplots(figsize=(9, 4))
ax.loglog(sample_sizes, [t*1000 for t in times['linear']],
          's--', color='#1D4ED8', lw=2, ms=8, label='SVM Linéaire')
ax.loglog(sample_sizes, [t*1000 for t in times['rbf']],
          'o-',  color='#D97706', lw=2, ms=8, label='SVM RBF')
ax.set_xlabel('Taille du dataset n (log)', fontsize=11)
ax.set_ylabel('Temps d\'entraînement (ms, log)', fontsize=11)
ax.set_title(
    'Scalabilité : SVM Linéaire vs RBF\n'
    'Complexité théorique SVM : O(n²) à O(n³) selon l\'implémentation',
    fontsize=11, fontweight='bold')
ax.legend(fontsize=10)
plt.tight_layout()
plt.savefig('tp_fig15_scalabilite.png', dpi=150, bbox_inches='tight')
plt.show()

print('\n📌 Limite majeure des SVM à noyau :')
print('   - Complexité en temps  : O(n² × d) à O(n³)')
print('   - Complexité en espace : O(n²) pour stocker la matrice de Gram')
print('   → Pour n > 100 000 : préférer SGD, Random Forests, ou réseaux de neurones')
```

## Partie 5 — Synthèse finale et tableau de bord

### Dashboard récapitulatif : tous les modèles testés

```python
# ── Dashboard final — tous les modèles sur les 3 datasets ─────────────────────
all_models = [
    ('SVM Linéaire',          SVC(kernel='linear', C=1.0)),
    ('SVM Poly d=2',          SVC(kernel='poly',   degree=2, coef0=1, C=1.0)),
    ('SVM Poly d=3',          SVC(kernel='poly',   degree=3, coef0=1, C=1.0)),
    ('SVM Poly d=4',          SVC(kernel='poly',   degree=4, coef0=1, C=1.0)),
    ('SVM RBF γ=0.1  C=1',   SVC(kernel='rbf', gamma=0.1,   C=1.0)),
    ('SVM RBF γ=0.8  C=1',   SVC(kernel='rbf', gamma=0.8,   C=1.0)),
    ('SVM RBF γ=2    C=10',  SVC(kernel='rbf', gamma=2.0,   C=10.0)),
    ('SVM RBF γ=20   C=1000',SVC(kernel='rbf', gamma=20.0,  C=1000.0)),
    ('SVM Sigmoïde',          SVC(kernel='sigmoid', gamma=0.3, coef0=0)),
]

dashboard = {dname: {} for dname in dataset_names}
for mname, clf in all_models:
    for dname in dataset_names:
        d = scaled[dname]
        clf_c = SVC(**clf.get_params())
        clf_c.fit(d['Xtr'], d['ytr'])
        acc_te = clf_c.score(d['Xte'], d['yte'])
        dashboard[dname][mname] = acc_te

# Heatmap du dashboard
model_names = [n for n,_ in all_models]
mat = np.array([[dashboard[dn][mn] for dn in dataset_names]
                for mn in model_names])

fig, ax = plt.subplots(figsize=(10, 9))
im = ax.imshow(mat, cmap='RdYlGn', vmin=0.50, vmax=1.0, aspect='auto')
cbar = plt.colorbar(im, ax=ax)
cbar.set_label('Accuracy Test', fontsize=11)

ax.set_xticks(range(len(dataset_names))); ax.set_xticklabels(dataset_names, fontsize=12)
ax.set_yticks(range(len(model_names)));   ax.set_yticklabels(model_names,   fontsize=10)
ax.set_xlabel('Dataset', fontsize=12); ax.set_title(
    'Dashboard final — Accuracy TEST de tous les modèles\n'
    'Vert = bonne accuracy | Rouge = mauvaise accuracy',
    fontsize=12, fontweight='bold')

for i, mn in enumerate(model_names):
    for j, dn in enumerate(dataset_names):
        v = mat[i,j]
        txt_col = 'white' if v > 0.94 or v < 0.65 else 'black'
        ax.text(j, i, f'{v:.0%}', ha='center', va='center',
                fontsize=11, fontweight='bold', color=txt_col)

plt.tight_layout()
plt.savefig('tp_fig16_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── Tableau de synthèse critique ──────────────────────────────────────────────
print('\n' + '='*80)
print('SYNTHÈSE FINALE — MÉTHODES À NOYAU')
print('='*80)

synthese = [
    ('NOYAU LINÉAIRE',
     'Pas de mapping — SVM classique',
     'Données linéairement séparables, grande dimension (d >> n)',
     'Paramètre : C uniquement',
     'Rapide, scalable, pas de surapprentissage lié au noyau'),

    ('NOYAU POLYNOMIAL',
     'K(x,x\') = (x·x\' + c)^d',
     'Relations polynomiales connues, NLP (n-grams), vision',
     'Paramètres : d (degré), c, C',
     'd trop élevé → surapprentissage. d impair → asymétrie'),

    ('NOYAU RBF (GAUSSIEN)',
     'K(x,x\') = exp(-γ||x-x\'||²)',
     'Cas général, structure inconnue, dimension faible',
     'Paramètres : γ, C — toujours faire GridSearch',
     'Ne scale pas bien (O(n²-n³)). Sensible à γ et C'),

    ('NOYAU SIGMOÏDE',
     'K(x,x\') = tanh(κ·x·x\' + θ)',
     'Analogie avec réseau de neurones (couche cachée)',
     'Paramètres : κ, θ — attention : pas toujours SDP',
     'Non garanti valide (Mercer) → résultats instables'),
]

for name, formula, usage, params, limite in synthese:
    print(f'\n  ▶ {name}')
    print(f'    Formule  : {formula}')
    print(f'    Usage    : {usage}')
    print(f'    Params   : {params}')
    print(f'    Limite   : {limite}')

print('\n' + '='*80)
print('RÈGLES PRATIQUES')
print('='*80)
print('''
  1. TOUJOURS normaliser les données avant d'utiliser un noyau RBF ou polynomial.
     → Sans normalisation, les distances et produits scalaires sont faussés.

  2. TOUJOURS utiliser GridSearchCV avec validation croisée.
     → Évaluer sur le test set sans cross-validation = estimation biaisée.

  3. Diagnostiquer avec les courbes d'apprentissage :
     - Deux courbes basses et proches → sous-apprentissage → augmenter C ou γ
     - Grand gap persistant           → surapprentissage   → diminuer C ou γ
     - Deux courbes hautes et proches → bon modèle

  4. Si n > 50 000 : le SVM à noyau devient trop lent.
     → Alternatives : Nystroem Approximation, Random Kitchen Sinks,
        SGD avec kernel trick approché, ou réseaux de neurones.

  5. Le noyau RBF est le point de départ par défaut pour les données non linéaires,
     mais ce n'est pas universellement optimal.
     → Tester systématiquement plusieurs noyaux sur les données réelles.
''')
print('='*80)
print('\n✅ TP terminé — toutes les figures sont sauvegardées.')
```
